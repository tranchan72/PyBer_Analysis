# Import dependencies

```python
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

# Create a plot using the object-oriented interface method

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis, label='Boston', color = "green", marker = "D", linewidth = 2)
# plot(x, y, color='green', marker='o', linestyle='dashed', linewidth=2, markersize=12)
ax.set_xlabel('Date')
ax.set_ylabel('Fare($)')
ax.set_title("PyBer Fare by Month")
ax.set_ylim(0, 45)
ax.legend()
ax.grid()
```

# Make bubles bigger

```python
y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*3)
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

# Using the list comprehension technique to multiply each data point in the y-axis list

```python
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

# Create a Scatter Plot Using the Object-Oriented Interface

```python
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s = [i * 3 for i in y_axis],label='Boston', 
           facecolors = "skyblue", edgecolors='black', linewidth=2, alpha=0.2)
ax.invert_yaxis()
ax.set_xlim(0, 45)
ax.set_xlabel('Fare($)')
ax.set_ylabel('Date')
ax.set_title("PyBer Fare by Month")
ax.legend()
```

# Create a Pie Chart Using the MATLAB Method

```python
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
# Pop out the biggest value
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
plt.show()
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", 
          "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')
plt.show()
```

# Create a Pie Chart Using the Object-Oriented Interface

```python
fig, ax = plt.subplots(figsize =(8, 8))
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", 
          "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
ax.pie(y_axis,
        explode = explode_values,
        colors = colors,
        labels = x_axis,
        autopct = '%.1f%%',
        shadow = True,
        counterclock = False,
        startangle = 90,
      )
plt.show()
```
